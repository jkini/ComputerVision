```python
import pandas as pd
import cv2 as ocr
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df_train = pd.DataFrame.from_csv('pairsDevTrain.csv').reset_index()
dataFolder = 'images-lfw/'
```

```python
im1 = df_train.iloc[0]['name1']+"/"+df_train.iloc[0]['name1']+'_'+str(df_train.iloc[0]['im1']).zfill(4)+".jpg"
im2 = df_train.iloc[0]['name2']+"/"+df_train.iloc[0]['name2']+'_'+str(df_train.iloc[0]['im2']).zfill(4)+".jpg"
df_train.iloc[0]['label']
```

```python
im_left = ocr.imread(dataFolder+im1,0)
im_left = im_left.astype('float32')/255.0
im_right = ocr.imread(dataFolder+im2,0)
im_right = im_right.astype('float32')/255.0
im_right.shape
```

```python
plt.imshow(im_left)
```

```python
import os
from PIL import Image
for dirname in os.listdir('images-lfw/'):
    for imname in os.listdir('images-lfw/'+dirname):
        img = Image.open('images-lfw/'+dirname+'/'+imname)
        new_img = img.resize((32,32))
        if not os.path.exists('images-32-lfw/'+dirname+'/'):
            os.makedirs('images-32-lfw/'+dirname+'/')
        new_img.save('images-32-lfw/'+dirname+'/'+imname, "JPEG", optimize=True)
```
